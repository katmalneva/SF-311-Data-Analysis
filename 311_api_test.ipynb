```python
import pandas as pd
from sodapy import Socrata
import datetime as dt
import requests
```

```python
#method 1
client = Socrata("data.sfgov.org", None)
results = client.get("vw6y-z8j6", limit=10)
df = pd.DataFrame.from_records(results)
df.head()
```

```python
#date to search from then to now for new data pull currently set to two days ago
query_date = str(dt.date.today() - dt.timedelta(days= 2))
```

```python
query_date
```

```python
#method 2 - actually works
url = f"https://data.sfgov.org/resource/vw6y-z8j6.json?$where=updated_datetime>='{query_date}'&$limit=999999"
response = requests.get(url)
collection = response.json()
df = pd.DataFrame(collection)
df['requested_datetime'] = pd.to_datetime(df['requested_datetime'])
df.sort_values(by='requested_datetime', inplace=True, ascending=True)
df.head()
```
